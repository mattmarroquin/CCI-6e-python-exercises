```python
def zero_matrix(matrix):
    column_length = len(matrix)  
    row_length = len(matrix[0])
    rows = [False] * column_length
    cols = [False] * row_length
    
    # M > N
    if column_length > row_length: 
        for r in range(column_length):
            for c in range(row_length):
                if matrix[r][c] == 0:
                    rows[r] = True
                    cols[c] = True
    # M < N
    elif row_length > column_length: 
        for r in range(column_length):
            for c in range(row_length):
                if matrix[r][c] == 0:
                    rows[r] = True
                    cols[c] = True
    # M = N
    else:
        for c in range(column_length):
            for r in range(row_length):
                if matrix[r][c] == 0:
                    rows[r] = True
                    cols[c] = True
    
    #zero rows
    row_num = 0
    for row in rows:
        if row:
            for i in range(row_length):
                matrix[row_num][i] = 0
        row_num += 1
    
    #zero columns
    col_num = 0
    for col in cols:
        if col:
            for i in range(column_length):
                matrix[i][col_num] = 0
        col_num += 1
    
    return matrix
```
